```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as nd
import seaborn as sns

delivery_events = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/delivery_events.csv")
loads = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/loads.csv")
trips = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/trips.csv")
routes = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/routes.csv")
```

```python
events_cols = ['load_id', 'trip_id', 'event_type', 'facility_id','actual_datetime', 'detention_minutes',
       'on_time_flag', 'location_city']
loads_cols = ["load_id", "route_id", "revenue", "fuel_surcharge", "accessorial_charges"]
trips_cols = ["trip_id","actual_distance_miles", "fuel_gallons_used", "actual_duration_hours", "idle_time_hours"]
routes_cols = ["route_id", "origin_city", "destination_city"]

loads_and_trips = delivery_events[events_cols].merge(loads[loads_cols], on="load_id", how="inner")
loads_and_trips = loads_and_trips.merge(trips[trips_cols], on="trip_id", how="inner")
base_table = loads_and_trips.merge(routes[routes_cols], on="route_id", how="inner")

print(base_table.columns)
```

```python
# 1列ずつ個別に datetime 型に変換して上書きする
base_table["actual_datetime"] = pd.to_datetime(base_table["actual_datetime"])

base_table["actual_datetime"] = base_table["actual_datetime"].dt.date
```

```python
base_table["total_sales"] = base_table["revenue"] + base_table["fuel_surcharge"] + base_table["accessorial_charges"]
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 1. 日付列を確実に datetime 型に変換
base_table["actual_datetime"] = pd.to_datetime(
    base_table["actual_datetime"], errors="coerce"
)

# 2. 日付の古い順にソート
df_sorted = base_table.dropna(subset=["actual_datetime"]).sort_values(
    "actual_datetime"
)

# 3. 1日ごとの売上合計に集約
df_sorted["date"] = df_sorted["actual_datetime"].dt.date
daily_sales = df_sorted.groupby("date")["total_sales"].sum()

# 4. 7日・30日の移動平均を計算
rolling_7 = daily_sales.rolling(window=7, min_periods=1).mean()
rolling_30 = daily_sales.rolling(window=30, min_periods=1).mean()

# 5. グラフ描画（🚨エラーを起こす怪しいパラメータはすべて削除しました！）
plt.figure(figsize=(12, 5))

# 各データをプロット
plt.plot(
    daily_sales.index,
    daily_sales,
    label="Actual (Daily Total)",
    alpha=0.4,
    color="blue",
)
plt.plot(
    rolling_7.index,
    rolling_7,
    label="7-day Rolling Mean",
    linewidth=2,
    color="orange",
)
plt.plot(
    rolling_30.index,
    rolling_30,
    label="30-day Rolling Mean",
    linewidth=3,
    color="green",
)

# 💡Colab環境で100%エラーを避けるため、タイトル類を英語表記にしています
plt.title("Total Sales Rolling Mean & Time Series Analysis")
plt.xlabel("Date")
plt.ylabel("Sales Amount")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)

# 横軸の日付が見やすくなるように斜めにする設定
plt.gcf().autofmt_xdate()

plt.tight_layout()
plt.show()
```

```python
from os import rename
#燃料費算出と結合
fuel_price = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/fuel_purchases.csv")
fuel_cost = fuel_price.groupby("trip_id")["total_cost"].sum().reset_index()

fuel_cost.rename(columns={"total_cost": "total_fuel_cost"}, inplace=True)

fuel_col = ["trip_id", "total_fuel_cost"]
base_table = base_table.merge(fuel_cost[fuel_col],on="trip_id",how="inner")
```

```python
#回転効率
base_table["total_revenue"] = (
    base_table["revenue"]
    + base_table["fuel_surcharge"]
    + base_table["accessorial_charges"]
)

base_table["fuel_per_mile"] = (
    base_table["fuel_gallons_used"]
    / base_table["actual_distance_miles"]
)

base_table["idle_ratio"] = (
    base_table["idle_time_hours"]
    / base_table["actual_duration_hours"]
)
# Calculate profit as revenue - fuel costs since we don't have more costs.
base_table["profit"] = base_table["total_revenue"] - base_table['total_fuel_cost']
base_table['profitability'] = base_table['profit']<0
```

```python
#  Group by Route
route_summary = base_table.groupby(by='route_id').agg({
    "profit": 'mean',
    'profitability':'mean',
    'fuel_per_mile' : 'mean',
    'actual_duration_hours': 'mean',
    'trip_id':'count'
    })

route_summary = route_summary.rename(columns={
    "profit": "avg_profit",
    'profitability':'pct_unprofitable',
    'fuel_per_mile' : 'avg_fuel_per_mile',
    'actual_duration_hours': 'avg_duration',
    'trip_id':'total_trips'})

route_summary['pct_unprofitable'] = route_summary['pct_unprofitable'] *100

route_summary= route_summary.sort_values('pct_unprofitable', ascending = False)
route_summary["profit_per_hour"] = route_summary["avg_profit"] / route_summary["avg_duration"]

print(route_summary[['avg_profit','profit_per_hour']].head(6).reset_index())
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))

# 1.散布図を描画
sns.scatterplot(
    data=except_worstroute_summery,
    x="avg_duration",
    y="profit_per_hour"
)

# 2.対数回帰曲線の計算
x_data = except_worstroute_summery["avg_duration"]
y_data = except_worstroute_summery["profit_per_hour"]

# xを対数（log）に変換して1次フィットさせることで、y = a * ln(x) + b の係数を算出します
a, b = np.polyfit(np.log(x_data), y_data, 1)
curve_equation = f"y = {a:.2f} * ln(x) + {b:.2f}"
print(f"算出した回帰曲線式: {curve_equation}")

# 3.曲線を描くための滑らかなX軸のデータを作成してプロット
x_smooth = np.linspace(x_data.min(), x_data.max(), 200)
y_smooth = a * np.log(x_smooth) + b

plt.plot(
    x_smooth, y_smooth,
    color="crimson",
    linewidth=2.5,
    label="Logarithmic Regression Curve"
)

# 4. グラフ内に回帰曲線の数式をテキストボックスで表示
x_pos = x_data.max() - ((x_data.max() - x_data.min()) * 0.45)
y_pos = y_data.min() + ((y_data.max() - y_data.min()) * 0.05)

plt.text(
    x_pos, y_pos,
    f"Regression Curve:\n{curve_equation}",
    fontsize=12,
    color="crimson",
    weight="bold",
    bbox=dict(facecolor='white', alpha=0.8, edgecolor='crimson', boxstyle='round,pad=0.5')
)

x_max = x_data.max()
plt.xticks(np.arange(0, x_max + 5, 5))

# 5. タイトルと仕上げ
plt.title("Duration vs Profit per Hour with Regression Curve (Except Worst Routes)")
plt.xlabel("Average Duration (hours)")
plt.ylabel("Profit per Hour ($)")
plt.legend(loc="upper left")
plt.grid(True, linestyle=":", alpha=0.6)

plt.show()
```

```python
import numpy as np
import pandas as pd

# 1. 回帰曲線の係数 (a, b) から合格ラインを計算
x_data = except_worstroute_summery["avg_duration"]
y_data = except_worstroute_summery["profit_per_hour"]
a, b = np.polyfit(np.log(x_data), y_data, 1)
except_worstroute_summery["target_profit_per_hour"] = a * np.log(x_data) + b

# 2. 曲線の下にあるかどうかのフラグ（判定）をデータに追加
except_worstroute_summery["is_below_curve"] = (
    except_worstroute_summery["profit_per_hour"] < except_worstroute_summery["target_profit_per_hour"]
)

# 3. 5時間刻みのグループ（ビン）を定義
max_hour = int(np.ceil(x_data.max() / 5) * 5)
bins = np.arange(0, max_hour + 5, 5)
labels = [f"{bins[i]}-{bins[i+1]}h" for i in range(len(bins)-1)]

# 4. すべてのルートを5時間刻みのグループに分類
except_worstroute_summery["duration_group"] = pd.cut(
    except_worstroute_summery["avg_duration"],
    bins=bins,
    labels=labels,
    include_lowest=True
)

# 5. グループごとに「総個数」と「下にある個数」を同時に集計
grouped = except_worstroute_summery.groupby("duration_group", observed=False).agg(
    総ルート数=("is_below_curve", "count"),
    曲線の下にあるルート数=("is_below_curve", "sum")
)

# 6.割合（％）を計算する
grouped["改善対象の割合 (%)"] = (grouped["曲線の下にあるルート数"] / grouped["総ルート数"] * 100).fillna(0)

# 表示用にインデックスを列に戻して整理
summary_table = grouped.reset_index().rename(columns={"duration_group": "運行時間帯"})

print("===== 運行時間帯（5h刻み）ごとの改善対象の割合 =====")
print(summary_table.to_string(index=False, formatters={"改善対象の割合 (%)": "{:.1f}%".format}))
```

```python
#運行時間15時間以下のコースに15時間運行分の最低保証料金の導入(赤字コース・黒字コース関係なく)
import math

route_summary["total_hour"] = base_table.groupby(by='route_id')['actual_duration_hours'].sum()

price_increase = route_summary[route_summary["avg_duration"] < 15]
control_group = route_summary[route_summary["avg_duration"] >= 15]

minimum_guarantee = 60.47 * math.log(15) + -111.78

price_increase["profit_per_hour"] = minimum_guarantee

price_increase_profit = price_increase["profit_per_hour"] * price_increase["total_hour"]
control_group_profit = control_group["profit_per_hour"] * control_group["total_hour"]

total_profit = price_increase_profit.sum() + control_group_profit.sum()

increase_profit =total_profit/(base_table["profit"].sum())*100 - 100

print(f"利益がマイナスのコースを黒字コースの下限25％に合わせた場合の利益上昇幅は{increase_profit}")
```

```python
import numpy as np
import pandas as pd

# 1. データのコピーを作成
sim_df = except_worstroute_summery.copy()

# 2. 各ルートの現在の「回帰曲線上の合格ライン(y値)」を計算
sim_df["curve_y"] = a * np.log(sim_df["avg_duration"]) + b

# 3. 各ルートの「総運行時間(total_hour)」と「現在の総利益(total_profit)」を確保
# ※前段の集計で入っているはずですが、万が一のために再計算・マージを担保します
if "total_hour" not in sim_df.columns:
    # 以前作成した worstroute_summery のロジックと同様に元のdfから総時間を引っ張るか、
    # なければ「平均時間 × 運行回数」で総運行時間を算出します
    sim_df["total_hour"] = sim_df["avg_duration"] * sim_df["total_trips"]

if "total_profit" not in sim_df.columns:
    sim_df["total_profit"] = sim_df["avg_profit"] * sim_df["total_trips"]


# 4. 🔥 【重要】条件分岐ロジックを関数として定義
def calculate_new_profit_per_hour(row):
    # 条件A: 15時間以上 料金かつ 回帰曲線よりも下の点
    if row["avg_duration"] >= 15 and row["profit_per_hour"] < row["curve_y"]:

        # 現行の時間あたり利益を10%アップさせた値
        pot_15pct_up = row["profit_per_hour"] * 1.15

        # もし10%アップさせた値が、回帰曲線のy値を超える、あるいは「その差が現行の10%以内」の場合
        # つまり、10%値上げすれば回帰曲線に届く（または超える）場合は、回帰曲線の値(curve_y)を採用
        if pot_15pct_up >= row["curve_y"]:
            return row["curve_y"]
        else:
            # それ以外（まだ回帰曲線まで遠い）なら、きっちり10%値上げ
            return pot_15pct_up

    else:
        # 条件に合わないルート（15h未満、または最初から合格ライン以上のルート）は現状維持
        return row["profit_per_hour"]

# 5. 定義したロジックをデータフレーム全体に適用
sim_df["new_profit_per_hour"] = sim_df.apply(calculate_new_profit_per_hour, axis=1)

# 6. 改善後の新しい総利益を計算
sim_df["new_total_profit"] = sim_df["new_profit_per_hour"] * sim_df["total_hour"]

# 7. 🏁 【効果検証】会社全体の総利益に対して何％向上するかをシミュレーション
original_total_profit = base_table["profit"].sum()  # 1枚目の元の総利益（分母）

# ワースト5を削って生まれた空き時間のリプレイス利益（前回の improvement）
# もし今回のシミュレーションに「ワースト5の時間リプレイス」も合わせて乗せる場合は足してください。
# ここでは純粋に「残ったルート（except_worstroute）に対する今回の値上げ効果」を計算します。
actual_previous_profit = except_worstroute["profit"].sum()
new_projected_profit = sim_df["new_total_profit"].sum()

# 利益の純増額
net_increase = new_projected_profit - actual_previous_profit
final_growth_rate = (net_increase / original_total_profit) * 100

print("===== 📈 15時間以上・曲線以下のルート 10%値上げ試算結果 =====")
print(f"・会社全体の元々の総利益: {original_total_profit:,.0f} ドル")
print(f"・この局所値上げ戦略による利益純増額: +{net_increase:,.0f} ドル")
print(f"・📊 【結論】会社全体の利益は 【 {final_growth_rate:.2f}% 】 向上します！")
print("==========================================================")

# 対象となったルートがどれくらい値上げされたか上位5件を確認
targets_preview = sim_df[sim_df["profit_per_hour"] != sim_df["new_profit_per_hour"]]
print("\n💡 値上げ対象となったルートの確認（一部）:")
print(targets_preview[["avg_duration", "profit_per_hour", "curve_y", "new_profit_per_hour"]].head())
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns  # seaborn

# 1. 相関係数の算出
correlation = drivers_summery["profitability"].corr(
    drivers_summery["operation_rate"]
)
print(f"算出された相関係数: {correlation:.4f}")

# 2. 軸の割り当てに沿って正しく閾値を計算 (x=operation_rate, y=profitability)
x_lower_limit = drivers_summery["operation_rate"].quantile(0.25)
y_lower_limit = drivers_summery["profitability"].quantile(0.25)

print(f"operation_rateの下位25%閾値(x軸の縦線): {x_lower_limit:.4f}")
print(f"profitabilityの下位25%閾値(y軸の横線): {y_lower_limit:.4f}")

# 3. 回帰式の計算 (傾き a と 切片 b を算出)
a, b = np.polyfit(drivers_summery["operation_rate"], drivers_summery["profitability"], 1)
regression_equation = f"y = {a:.4f}x + {b:.4f}"
print(f"算出した回帰式: {regression_equation}")

# 4. グラフ描画
plt.figure(figsize=(10, 6))

sns.regplot(
    data=drivers_summery,
    x="operation_rate",
    y="profitability"
)

# 5. 正しい位置に基準線を引く
plt.axvline(
    x=x_lower_limit,
    color="red",
    linestyle="--",
    linewidth=1.5,
    label=f"operation_rate Lower 0.25 ({x_lower_limit:.4f})",
)
plt.axhline(
    y=y_lower_limit,
    color="red",
    linestyle="--",
    linewidth=1.5,
    label=f"profitability Lower 0.25 ({y_lower_limit:.4f})",
)

# 🔥 6. 【修正部分】エラーが出ない方法で最大値・最小値の幅（レンジ）を計算
x_max = drivers_summery["operation_rate"].max()
x_min = drivers_summery["operation_rate"].min()
x_range = x_max - x_min  # これがptpの代わりになります

y_max = drivers_summery["profitability"].max()
y_min = drivers_summery["profitability"].min()
y_range = y_max - y_min  # これがptpの代わりになります

# テキストを表示する座標を計算（右下の安全圏）
x_pos = x_max - (x_range * 0.4)
y_pos = y_min + (y_range * 0.1)

# グラフ内に回帰式を表示
plt.text(
    x_pos, y_pos,
    f"Regression Equation:\n{regression_equation}",
    fontsize=12,
    color="blue",
    weight="bold",
    bbox=dict(facecolor='white', alpha=0.8, edgecolor='blue', boxstyle='round,pad=0.5')
)

# 7. タイトルと凡例の整理
plt.title(
    f"Scatter Plot with Regression Line (Correlation: {correlation:.2f})"
)
plt.legend(loc="upper left")
plt.grid(True, linestyle=":", alpha=0.6)

plt.show()
```
